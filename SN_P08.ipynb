## Introducción
Fuzzy clusturing

Agrupamiento difuso.

El agrupamiento de datos numéricos consiste en a partir de uns seríe o conjunto de datos poder clasificarlos y modelarlos matemáticamente. Teniendo como propósito identificar grupos naturales de datos a partir de un conjunto grande de datos para producir una concisa representación del comportamiento del sistema.

La tootbox de matlab te permite encontrar el agrupamiento de datos. Pueden ser usados para general fuzzificacion por medio de sugeno

FCM.

Es una técnica de agrupación de datos donde cada punto de de esto pertenece o tiene agrupamiento con cierto grado el cuál es expresado por una función de membresía la cuál determina el grado para cada uno de ellas

La línea de comando FCM empieza con una condicional inicial la cuál nos dicta el centro de agrupación de los datos los cuales están identados para marcar la iniciación principal para cada agrupación.

La mayoría de las veces está inicialización es incorrecto pues solo nos brinda un espacio inicial de trabajo. FCM trabajo agrupando cada dato de manera iterativa, 
FCM mueve los centros de agrupación de datos a los valores correctos. La iteracion está Dada tratando de minimizar la distancia entre cada punto para ajustar la distancia al centro de agrupación.

Este comando nos arroja una lista de centros de agrupación y un número muy grande de grados de membresía para cada punto de información. Se puede usar esta información de regreso para crear sistemas interferencia difusa dando las funciones de membresía lo cual representará las cualidades de fuzzificacion de cada agrupación.


## Problema 1
Implemente tres funciones, para calcular las siguientes funciones de membresía.
\begin{align}
        \text{μ}_1(x) = \left. \frac{1}{1+k(x-m)^2}
        \right. k > 1
    \end{align}
;
\begin{align}
        \text{μ}_2(x) = \left\{
        \begin{array}{cl}
        0 & x \le a \\
        \frac{k(x-a)^2}{1+k(x-a)^2} & x > a.
        \end{array}
        \right. k > 1
    \end{align}
;
\begin{align}
        \text{μ}_3(x) = \left\{
        \begin{array}{cl}
        0 & x \neq m \\
        1 & x = m.
        \end{array}
        \right.
    \end{align}


Los párametros que no muestran restricción pueden tener cualquier número real como argumento, la función debe ser capaz de evaluar vectores de cualquier dimensión. La función μ3 recibe el nombre de singleton.
Grafique las funciones para los siguientes parámetros (con las herramientas ya construidas)


1.   ${μ}_1(x)$ con m = 1, k =2
2.   ${μ}_1(x)$ con m = 2, k = 4
3.   ${μ}_2(x)$ con a = 1, k = 2
4.   ${μ}_2(x)$ con a = 2, k = 4
5.   ${μ}_3(x)$ con m = 2

```python
! pip install -U fuzzylab 
! pip install -U scikit-fuzzy
! pip install simpful

import numpy as np
import matplotlib.pyplot as plt

#funcion para plotear
def plotear(x,y,title)->None:
  plt.figure(num = 0,dpi=120)
  plt.plot(x,y)
  plt.grid()
  plt.title(title)
  return None

#funcion para u1
def u1 (m,k)->None:
  x  = np.linspace(m-5,m+5,100)
  y = 1/(1+k*(x-m)**2)
  stitle = '${μ}_1(x)$' + ' m = {0} k = {1}'.format(m,k)
  plotear(x,y,stitle)
  return None

#funcion para u2
def u2 (a,k)->None:
  x = np.linspace(a-2,a+8,100)
  y = list()
  for value in x:
    if(value<=a):
      y.append(0)
    else:
      y.append((k*(value-a)**2)/(1+k*(value-a)**2))
  y = np.array(y)
  stitle = '${μ}_2(x)$' + ' a = {0} k = {1}'.format(a,k)
  plotear(x,y,stitle)
  return None

#funcion para u3
def u3 (m)->None:
  x  = np.linspace(m-5,m+5,101)
  y = list()
  for value in x:
    if(value!=m):
      y.append(0)
    else:
      y.append(1)
  
  y = np.array(y)
  stitle = '${μ}_3(x)$' + ' m = {0}'.format(m)
  plt.figure(num = 0,dpi=120)
  plt.plot(x,y,marker='o',linestyle='None')
  plt.grid()
  plt.title(stitle)
  return None
```

## Problema 2
Utilizando *fuzzylab*. Escoja tres conjuntos difusos A, B, C con dominio continuo. La elección debe ser de tal forma que la intersección de B con A y de B con C no sea vacia (esto es solo para visualizar la operación, y solo para eso).
Verifique las siguientes propiedades

1. $A \cup (B \cap C) = (A \cup B)\cap (A \cup C)$
2. $\overline{A \cap B}= \overline{A} \cup \overline{B}$
3. $\overline{A \cup B}= \overline{A} \cap \overline{B}$

Para la verificación, elabore la gráfica de los conjuntos de cada lado de la igualdad y posteriormente observe que el resultado gráfico es el mismo.

```python
import fuzzylab as fl
x = np.linspace(-11,5,101)
A = fl.trapmf(x,[-10, -8, -2, 2])
B = fl.trapmf(x,[-5, -3, 2, 4])
C = fl.trapmf(x,[-2, 0, 1, 3])

plt.figure(num = 0,dpi=120)
plt.plot(x,A,label = 'A')
plt.plot(x,B,label = 'B')
plt.plot(x,C,label = 'C')
plt.grid()
plt.legend()
plt.title('A, B, C')
```

Para la propiedad 1 tenemos: 

```python
#funcion para plotear los lados de la propiedad
def subplt(x_axis,y1_axis,y2_axis,tit,title1,title2):
  fig, axs = plt.subplots(1,2,dpi=120)
  fig.suptitle(tit)
  axs[0].grid()
  axs[0].set_title(title1)

  axs[1].grid()
  axs[1].set_title(title2)

  axs[0].plot(x_axis,y1_axis)
  axs[1].plot(x_axis,y2_axis)
  return None

#propiedad 1
L1 = np.zeros(shape=x.size)
D1 = np.zeros(shape=x.size)
for i,value in enumerate(x):
    #lado izq
    L1[i]=max(A[i],min(B[i],C[i]))
    #lado der
    D1[i]=min(max(A[i],B[i]),max(A[i],C[i]))

#plot de la propiedad
subplt(x,L1,D1,'Propiedad 1','$A \cup (B \cap C)$','$(A \cup B)\cap (A \cup C)$')
```

Para la propiedad 2:

```python
#propiedad 2
L2 = np.zeros(shape=x.size)
D2 = np.zeros(shape=x.size)
for i,value in enumerate(x):
    #lado izq
    L2[i]=1-min(A[i],B[i])
    #lado der
    D2[i]=max(1-A[i],1-B[i])

#plot de la propiedad
subplt(x,L2,D2,'Propiedad 2','$\overline{A \cap B}$','$\overline{A} \cup \overline{B}$')
```

Para la propiedad 3:

```python
#propiedad 3
L3 = np.zeros(shape=x.size)
D3 = np.zeros(shape=x.size)
for i,value in enumerate(x):
    #lado izq
    L3[i]=1-max(A[i],B[i])
    #lado der
    D3[i]=min(1-A[i],1-B[i])

#plot de la propiedad
subplt(x,L3,D3,'Propiedad 3','$\overline{A \cup B}$','$\overline{A} \cap \overline{B}$')
```

## Problema 3
En la documentación del módulo *simpful* puedes encontrar la implementación del problema de la propina. Resuelve el problema 1 de la P06 con este módulo

En la sigueinte sección de código se contruye un sistema de inferencia difuso tipo Mandani.
Se plantea un problema para calcular una propina en base a la experiencia de un cliente, calificando el servicio y la comida.
Se tienen diferentes rangos para la comida:
* Rancia
* Aceptable
* Rica
* Muy rica
 Para el servio:
* Malo
* Normal
* Bueno
* Muy bueno

Y para la salida (propina) se tiene:
* Muy pequeña
* Pequeña
* Media
* Generosa

Más adelante se muestran las gráficas correspondientes, así como un ejemplo.

```python
import simpful as sf

#variante de la propina
FS = sf.FuzzySystem()

#Define los conjuntos difusos y variables linguisticas
S_1 = sf.FuzzySet(function=sf.Triangular_MF(a=0,b=0,c=50),term = "malo")
S_2 = sf.FuzzySet(function=sf.Triangular_MF(a=0,b=50,c=100),term = "normal")
S_3 = sf.FuzzySet(function=sf.Triangular_MF(a=50,b=100,c=100),term = "bueno")
S_4 = sf.FuzzySet(function=sf.Triangular_MF(a=75,b=100,c=100),term = "muy_bueno")

SL = sf.LinguisticVariable([S_1,S_2,S_3,S_4], concept = "experiencia como consumidor", universe_of_discourse=[0,100])
FS.add_linguistic_variable("Servicio",SL)
SL.plot()

F_1 = sf.FuzzySet(function=sf.Triangular_MF(a=0,b=0,c=100),term = "rancia")
F_2 = sf.FuzzySet(function=sf.Triangular_MF(a=0,b=50,c=100),term = "aceptable")
F_3 = sf.FuzzySet(function=sf.Triangular_MF(a=50,b=100,c=100),term = "rica")
F_4 = sf.FuzzySet(function=sf.Triangular_MF(a=75,b=100,c=100),term = "muy_rica")
FL= sf.LinguisticVariable([F_1,F_2,F_3,F_4], concept = "experiencia como consumidor", universe_of_discourse=[0,100])
FS.add_linguistic_variable("Comida",FL)
FL.plot()
#Output
P_1 = sf.FuzzySet(function=sf.Triangular_MF(a=0,b=0,c=3),term = "muy_pequeña")
P_2 = sf.FuzzySet(function=sf.Triangular_MF(a=0,b=0,c=5),term = "pequeña")
P_3 = sf.FuzzySet(function=sf.Triangular_MF(a=0,b=7.5,c=15),term = "media")
P_4 = sf.FuzzySet(function=sf.Triangular_MF(a=10,b=15,c=15),term = "generosa")
PL = sf.LinguisticVariable([P_1,P_2,P_3,P_4], universe_of_discourse=[0,15])
PL.plot()

FS.add_linguistic_variable("Propina",PL)

#Fuzzy rules
R1 = "IF (Servicio IS malo) OR (Comida IS rancia) THEN (Propina IS muy_pequeña)"
R2 = "IF (Servicio IS normal) OR (Comida IS aceptable) THEN (Propina IS pequeña)"
R3 = "IF (Servicio IS bueno) OR (Comida IS rica) THEN (Propina IS media)"
R4 = "IF (Servicio IS muy_bueno) OR (Comida IS muy_rica) THEN (Propina IS generosa)"

FS.add_rules([R1, R2, R3,R4])

FS.set_variable("Servicio", 87)
FS.set_variable("Comida", 65)

# Mamdani inference
print(FS.Mamdani_inference(["Propina"]))


```

## Problema 4
En la documentación del módulo skfuzzy puedes encontrar la implementación del problema de la propina. Resuelve el problema 1 de la P06 con este módulo.
En la sigueinte sección de código se contruye un sistema de inferencia difuso tipo Mandani.

Se plantea un problema para calcular una propina en base a la experiencia de un cliente, calificando el servicio y la comida.
Se tienen diferentes rangos para la comida:
* Rancia
* Aceptable
* Rica
* Muy rica
 Para el servio:
* Malo
* Normal
* Bueno
* Muy bueno

Y para la salida (propina) se tiene:
* Muy pequeña
* Pequeña
* Media
* Generosa

Más adelante se muestran las gráficas correspondientes, así como un ejemplo.

```python
import skfuzzy as fuzz

#universos
x_com = np.arange(0, 101, 1)
x_serv = np.arange(0, 101, 1)
x_pro = np.arange(0, 16, 1)

#fuzzy membership functions

S_malo = fuzz.trimf(x_serv,[0,0,50])
S_norm = fuzz.trimf(x_serv,[0,50,100])
S_buen = fuzz.trimf(x_serv,[50,100,100])
S_mbue = fuzz.trimf(x_serv,[75,100,100])

C_ranc = fuzz.trimf(x_com,[0,0,50])
C_acep = fuzz.trimf(x_com,[0,50,100])
C_rica = fuzz.trimf(x_com,[50,100,100])
C_mric = fuzz.trimf(x_com,[75,100,100])

P_mpeq = fuzz.trimf(x_pro,[0,0,3])
P_pequ = fuzz.trimf(x_pro,[0,0,5])
P_medi = fuzz.trimf(x_pro,[0,7.5,15])
P_gene = fuzz.trimf(x_pro,[10,15,15])

#subplot para ver funciones
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 9))

ax0.plot(x_com, C_ranc, linewidth=1.5, label='Rancia')
ax0.plot(x_com, C_acep, linewidth=1.5, label='aceptable')
ax0.plot(x_com, C_rica, linewidth=1.5, label='rica')
ax0.plot(x_com, C_mric, linewidth=1.5, label='muy_rica')
ax0.set_title('Calidad de la comida')
ax0.legend()

ax1.plot(x_serv, S_malo, linewidth=1.5, label='malo')
ax1.plot(x_serv, S_norm, linewidth=1.5, label='normal')
ax1.plot(x_serv, S_buen, linewidth=1.5, label='bueno')
ax1.plot(x_serv, S_mbue, linewidth=1.5, label='muy_bueno')
ax1.set_title('Calidad del servicio')
ax1.legend()

ax2.plot(x_pro, P_mpeq, linewidth=1.5, label='muy_pequeña')
ax2.plot(x_pro, P_pequ, linewidth=1.5, label='pequeña')
ax2.plot(x_pro, P_medi, linewidth=1.5, label='media')
ax2.plot(x_pro, P_gene, linewidth=1.5, label='generosa')
ax2.set_title('Cantidad de propina')
ax2.legend()
#quitar bordes
for ax in (ax0, ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python
#Reglas, dependiendo de las entradas
cal_comida = 50
cal_servicio = 65

S_malo_ev = fuzz.interp_membership(x_serv,S_malo,cal_servicio)
S_norm_ev = fuzz.interp_membership(x_serv,S_norm,cal_servicio)
S_buen_ev = fuzz.interp_membership(x_serv,S_buen,cal_servicio)
S_mbue_ev = fuzz.interp_membership(x_serv,S_mbue,cal_servicio)

C_ranc_ev = fuzz.interp_membership(x_com,C_ranc,cal_comida)
C_acep_ev = fuzz.interp_membership(x_com,C_acep,cal_comida)
C_rica_ev = fuzz.interp_membership(x_com,C_rica,cal_comida)
C_mric_ev = fuzz.interp_membership(x_com,C_mric,cal_comida)

#regla 1 y activacion
regla1 = np.fmax(S_malo_ev,C_ranc_ev)
pro_act_mpeq = np.fmin(regla1,P_mpeq)
#regla 2
regla2 = np.fmax(S_norm_ev,C_acep_ev)
pro_act_pequ = np.fmin(regla2,P_pequ)
#regla 3
regla3 = np.fmax(S_buen_ev,C_rica_ev)
pro_act_medi = np.fmin(regla3,P_medi)
#regla 4
regla4 = np.fmax(S_mbue_ev,C_mric_ev)
pro_act_gene = np.fmin(regla4,P_gene)

#Defuzz
agregado = np.fmax(pro_act_mpeq,np.fmax(pro_act_pequ,np.fmax(pro_act_medi,pro_act_gene)))

propina = fuzz.defuzz(x_pro,agregado,'centroid')

print("La propina calculada es de: ")
print(propina)
```

## Problema 5
Reproduce el ejemplo: Fuzzy c-means clustering con skfuzzy. Se espera una explicación adicional a la proporcionada en la documentación como producto de la investigación realizada en la introducción.



```python
colors = ['b', 'orange', 'g', 'r', 'c', 'm', 'y', 'k', 'Brown', 'ForestGreen']

#Define los centros de los 3 cluster
centers = [
    [4,2],
    [1,7],
    [5,6]
]

sigmas = [
    [0.8,0.3],
    [0.3,0.5],
    [1.1,0.7]
]

np.random.seed(42)
xpts = np.zeros(1)
ypts = np.zeros(1)
labels = np.zeros(1)
for i, ((xmu,ymu),(xsigma,ysigma)) in enumerate(zip(centers,sigmas)):
  xpts = np.hstack((xpts, np.random.standard_normal(200) * xsigma + xmu))
  ypts = np.hstack((ypts, np.random.standard_normal(200) * ysigma + ymu))
  labels = np.hstack((labels, np.ones(200) * i))
fig0, ax0 = plt.subplots()
for label in range(3):
    ax0.plot(xpts[labels == label], ypts[labels == label], '.',
             color=colors[label])
ax0.set_title('Test data: 200 points x3 clusters.')
```

```python
# Set up the loop and plot
fig1, axes1 = plt.subplots(3, 3, figsize=(8, 8))
alldata = np.vstack((xpts, ypts))
fpcs = []

for ncenters, ax in enumerate(axes1.reshape(-1), 2):
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
        alldata, ncenters, 2, error=0.005, maxiter=1000, init=None)

    # Store fpc values for later
    fpcs.append(fpc)

    # Plot assigned clusters, for each data point in training set
    cluster_membership = np.argmax(u, axis=0)
    for j in range(ncenters):
        ax.plot(xpts[cluster_membership == j],
                ypts[cluster_membership == j], '.', color=colors[j])

    # Mark the center of each fuzzy cluster
    for pt in cntr:
        ax.plot(pt[0], pt[1], 'rs')

    ax.set_title('Centers = {0}; FPC = {1:.2f}'.format(ncenters, fpc))
    ax.axis('off')

fig1.tight_layout()
```

```python
fig2, ax2 = plt.subplots()
ax2.plot(np.r_[2:11], fpcs)
ax2.set_xlabel("Number of centers")
ax2.set_ylabel("Fuzzy partition coefficient")
```

Con el ejemplo: Fuzzy c-means clustering. Podemos ver como predicciones (clusters) de los datos obtenidos. Se puede intentar predecir los datos con mas centros (con mas puntos a predecir) para intentar conocer cual es la mejor representacion de los datos, por medio del FPC. El valor del FPC va de 0 a 1, este nos indica que tan bien representados estan nuestros datos por medio de los cluster.

```python
cntr, u_orig, _, _, _, _, _ = fuzz.cluster.cmeans(
    alldata, 3, 2, error=0.005, maxiter=1000)

# Show 3-cluster model
fig2, ax2 = plt.subplots()
ax2.set_title('Trained model')
for j in range(3):
    ax2.plot(alldata[0, u_orig.argmax(axis=0) == j],
             alldata[1, u_orig.argmax(axis=0) == j], 'o',
             label='series ' + str(j))
ax2.legend()
```

```python
# Generate uniformly sampled data spread across the range [0, 10] in x and y
newdata = np.random.uniform(0, 1, (1100, 2)) * 10

# Predict new cluster membership with `cmeans_predict` as well as
# `cntr` from the 3-cluster model
u, u0, d, jm, p, fpc = fuzz.cluster.cmeans_predict(
    newdata.T, cntr, 2, error=0.005, maxiter=1000)


cluster_membership = np.argmax(u, axis=0) 

fig3, ax3 = plt.subplots()
ax3.set_title('Random points classified according to known centers')
for j in range(3):
    ax3.plot(newdata[cluster_membership == j, 0],
             newdata[cluster_membership == j, 1], 'o',
             label='series ' + str(j))
ax3.legend()

plt.show()
```

Conociendo la mejor representación para los datos, podemos contruir un modelo con el número de cluster que mas se adapte, para despues realizar una predicción alrededor de los centros. Lo que nos permite clasificar los puntos de acuerdo al modelo generado.
